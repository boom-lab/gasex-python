```python
from gasex import sol,diff
import matplotlib.pyplot as plt
import numpy as np
```

```python
gases = ("He","Ne","N2","Ar","Kr","Xe","O2",'CH4')
ng = len(gases)
S = 35
T = range(0,30)
Sol = np.zeros(shape=(30,ng))
for i in range(ng):
    g = gases[i]
    Sol[:,i] = sol.sol_SP_pt(S,T,gas=g)
```

```python
plt.semilogy(T,Sol)
plt.xlabel("T ($^{\circ}$C)")
plt.ylabel("Solubility (mol $L^{-1}$)")
plt.show()
```

```python
Sc = np.zeros(shape=(30,ng))
for i in range(ng):
    g = gases[i]
    Sc[:,i] = diff.schmidt(S,T,gas=g)
plt.plot(T,Sc)
plt.xlabel("T ($^{\circ}$C)")
plt.ylabel("Schmidt number")
plt.show()
```
