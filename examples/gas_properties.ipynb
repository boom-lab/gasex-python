```python
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
gasex_path = os.path.abspath(os.path.join('..'))
# add path to main project directory
if gasex_path not in sys.path:
    sys.path.append(gasex_path)
from gasex import diff
from gasex import sol
```

```python
gases = ("He","Ne","N2","Ar","Kr","Xe","O2",'CH4')
ng = len(gases)
S = 35
T = range(0,30)
Sol = np.zeros(shape=(30,ng))
for i in range(ng):
    g = gases[i]
    Sol[:,i] = sol.sol_SP_pt(S,T,gas=g)
```

```python
plt.semilogy(T,Sol)
plt.xlabel("T ($^{\circ}$C)")
plt.ylabel("Solubility (mol $L^{-1}$)")
plt.show()
```

```python
Sc = np.zeros(shape=(30,ng))
for i in range(ng):
    g = gases[i]
    Sc[:,i] = diff.schmidt(S,T,gas=g)
plt.plot(T,Sc)
plt.xlabel("T ($^{\circ}$C)")
plt.ylabel("Schmidt number")
plt.show()
```
